## Guided Project: Visualizing The Gender Gap In College Degrees

The aim of this project is to practice storytelling with data visualization. It is not aim to commercial use or to mislead anyone.

Package using - pandas and matplotlib

```python
import pandas as pd
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
women_degrees_usa = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
women_degrees_usa.head()
```

```python
stem_category = ['Engineering',
             'Computer Science', 
             'Psychology', 
             'Biology', 
             'Physical Sciences', 
             'Math and Statistics'
            ]

fig = plt.figure(figsize=(18, 3))

for i in range(0,6):
    ax = fig.add_subplot(1,6,i+1)
    ax.plot(women_degrees_usa['Year'], women_degrees_usa[stem_category[i]], c='b', label='Women', linewidth=3)
    ax.plot(women_degrees_usa['Year'], 100-women_degrees_usa[stem_category[i]], c='orange', label='Men', linewidth=3)
    
    for key,spine in ax.spines.items(): #an axis spine is the data area boundary lines
        spine.set_visible(False)
    
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_category[i])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if i == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif i == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')

plt.show()
```

## Comparing across all degrees

```python
stem_category = ['Engineering',
             'Computer Science', 
             'Psychology', 
             'Biology', 
             'Physical Sciences', 
             'Math and Statistics'
            ]
lib_arts_category = ['Foreign Languages', 
                 'English', 
                 'Communications and Journalism', 
                 'Art and Performance', 
                 'Social Sciences and History'
                ]
other_category = ['Health Professions', 
              'Public Administration', 
              'Education', 
              'Agriculture',
              'Business', 
              'Architecture'
             ]
```

```python
fig = plt.figure(figsize=(16, 20))

## Generate first column of line charts. STEM degrees.
for i in range(0,18,3):
    cat_index = int(i/3)
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees_usa['Year'], women_degrees_usa[stem_category[cat_index]], c='b', label='Women', linewidth=3)
    ax.plot(women_degrees_usa['Year'], 100-women_degrees_usa[stem_category[cat_index]], c='orange', label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_category[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if cat_index == 0:
        ax.text(2003, 85, 'Women')
        ax.text(2005, 10, 'Men')
    elif cat_index == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2003, 7, 'Women')


## Generate second column of line charts. Liberal arts degrees.
for i in range(1,16,3):
    cat_index = int((i-1)/3)
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees_usa['Year'], women_degrees_usa[lib_arts_category[cat_index]], c='b', label='Women', linewidth=3)
    ax.plot(women_degrees_usa['Year'], 100-women_degrees_usa[lib_arts_category[cat_index]], c='orange', label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_category[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if cat_index == 0:
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')

## Generate third column of line charts. Other degrees.
for i in range(2,20,3):
    cat_index = int((i-2)/3)
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees_usa['Year'], women_degrees_usa[other_category[cat_index]], c='b', label='Women', linewidth=3)
    ax.plot(women_degrees_usa['Year'], 100-women_degrees_usa[other_category[cat_index]], c='orange', label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_category[cat_index])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if cat_index == 0:
        ax.text(2003, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif cat_index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2003, 30, 'Women')
        
plt.show()
```
